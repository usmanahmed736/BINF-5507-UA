```python
#Assignment 3
#Note on use of AI: I used ChatGPT to help me identify and color outliers black for the make_blobs and make_circles datasets. Every time I tried, the outliers would be colored a different colour and it was hard to tell clearly what were outliers.

#Import packages and datasets
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN, KMeans, AgglomerativeClustering
from sklearn.datasets import make_moons, make_blobs, make_circles

#Dataset
X, y = make_moons(n_samples=300, noise=0.05, random_state=42)

```

```python
#Apply algorithms to make_moons
#DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)

# k-Means
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans.fit_predict(X)

# Agglomerative Clustering
agglo = AgglomerativeClustering(n_clusters=2)
agglo_labels = agglo.fit_predict(X)

#Plot
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

axes[0].scatter(X[:, 0], X[:, 1], c=dbscan_labels, cmap='viridis')
axes[0].set_title("DBSCAN Clustering")

axes[1].scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis')
axes[1].set_title("k-Means Clustering")

axes[2].scatter(X[:, 0], X[:, 1], c=agglo_labels, cmap='viridis')
axes[2].set_title("Agglomerative Clustering")

for ax in axes:
    ax.set_xlabel("Feature 1")
    ax.set_ylabel("Feature 2")

plt.tight_layout()
plt.show()
```

```python
#Apply algorithms to make_blobs
#Dataset
X2, y2 = make_blobs(
    n_samples=300,
    centers=3,
    cluster_std=[1.0, 2.5, 0.5],
    random_state=42)

#DBSCAN
dbscan2 = DBSCAN(eps=0.8, min_samples=5)
dbscan_labels2 = dbscan2.fit_predict(X2)

#k-Means
kmeans2 = KMeans(n_clusters=3, random_state=42)
kmeans_labels2 = kmeans2.fit_predict(X2)

#Agglomerative Clustering
agglo2 = AgglomerativeClustering(n_clusters=3)
agglo_labels2 = agglo2.fit_predict(X2)

#Plot
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

axes[0].scatter(X2[:, 0], X2[:, 1], c=dbscan_labels2, cmap='viridis')
outliers = dbscan_labels2 == -1 #Points that don't fit in any cluster will be labelled -1, stored in variable outliers
axes[0].scatter(X2[outliers, 0], X2[outliers, 1], c='black', label='Outliers') #Used to colorcode outliers black for easy identification
axes[0].set_title("DBSCAN Clustering")
axes[0].legend(loc="best", fontsize=8)

axes[1].scatter(X2[:, 0], X2[:, 1], c=kmeans_labels2, cmap='viridis')
axes[1].set_title("k-Means Clustering")

axes[2].scatter(X2[:, 0], X2[:, 1], c=agglo_labels2, cmap='viridis')
axes[2].set_title("Agglomerative Clustering")

for ax in axes:
    ax.set_xlabel("Feature 1")
    ax.set_ylabel("Feature 2")

plt.tight_layout()
plt.show()
```

```python
#Apply Algorithms to make_circles
from sklearn.datasets import make_circles

#Dataset
X3, y3 = make_circles(n_samples=300, factor=0.5, noise=0.05, random_state=42)

#DBSCAN
dbscan3 = DBSCAN(eps=0.2, min_samples=5)
dbscan_labels3 = dbscan3.fit_predict(X3)


#k-Means
kmeans3 = KMeans(n_clusters=2, random_state=42)
kmeans_labels3 = kmeans3.fit_predict(X3)

#Agglomerative Clustering
agglo3 = AgglomerativeClustering(n_clusters=2)
agglo_labels3 = agglo3.fit_predict(X3)

#Plot
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

outliers3 = dbscan_labels3 == -1
axes[0].scatter(X3[~outliers3, 0], X3[~outliers3, 1], c=dbscan_labels3[~outliers3], cmap='viridis')
axes[0].scatter(X3[outliers3, 0], X3[outliers3, 1], c='black', label='Outliers')
axes[0].set_title("DBSCAN Clustering")
axes[0].legend()

axes[1].scatter(X3[:, 0], X3[:, 1], c=kmeans_labels3, cmap='viridis')
axes[1].set_title("k-Means Clustering")

axes[2].scatter(X3[:, 0], X3[:, 1], c=agglo_labels3, cmap='viridis')
axes[2].set_title("Agglomerative Clustering")

for ax in axes:
    ax.set_xlabel("Feature 1")
    ax.set_ylabel("Feature 2")

plt.tight_layout()
plt.show()
```
